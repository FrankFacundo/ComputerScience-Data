```python
import networkx as nx
import matplotlib.pyplot as plt


def synthesize_graph_data(num_nodes, probability):
    """
    Synthesize random graph data.

    Parameters:
    num_nodes (int): Number of nodes in the graph.
    probability (float): Probability of edge creation.

    Returns:
    G (networkx.Graph): A randomly generated graph.
    """
    # Create a random graph using the Erds-Rnyi model
    G = nx.erdos_renyi_graph(num_nodes, probability)
    return G


def plot_graph(G):
    """
    Plot the graph using networkx and matplotlib.

    Parameters:
    G (networkx.Graph): The graph to plot.
    """
    # Draw the graph
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G)  # positions for all nodes
    nx.draw(
        G,
        pos,
        with_labels=True,
        node_size=500,
        node_color="skyblue",
        font_size=10,
        font_color="black",
        edge_color="gray",
    )

    plt.title("Randomly Generated Graph")
    plt.show()


if __name__ == "__main__":
    # Parameters for graph synthesis
    num_nodes = 20  # number of nodes
    probability = 0.2  # probability of edge creation

    # Synthesize graph data
    G = synthesize_graph_data(num_nodes, probability)

    # Plot the graph
    plot_graph(G)
```

```python
num_nodes = 20  # number of nodes
probability = 0.2  # probability of edge creation

# Synthesize graph data
G = synthesize_graph_data(num_nodes, probability)
```

```python
import networkx as nx
from collections import Counter

EDGES = [
    ("A", "B"),
    ("A", "B"),
    ("A", "B"),
    ("B", "C"),
    ("A", "C"),
    ("C", "D"),
    ("A", "B"),
    ("B", "A"),
    ("B", "A"),
]

g = nx.DiGraph((x, y, {"weight": v}) for (x, y), v in Counter(EDGES).items())
print(*g.edges(data=True), sep="\n")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Suppose 'g' is your graph and it has weights on its edges
# If your graph does not have weights, you might first need to add them, for example:
# g.add_edge('a', 'b', weight=2)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(g)  # positions for all nodes

# Draw the nodes
nx.draw_networkx_nodes(g, pos, node_size=500, node_color="skyblue")

# Draw the edges
nx.draw_networkx_edges(g, pos, edge_color="gray")

# Draw the labels
nx.draw_networkx_labels(g, pos, font_size=10, font_color="black")

# Extract edge labels from graph
edge_labels = nx.get_edge_attributes(g, "weight")
nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels, label_pos=0.3)
plt.title("Randomly Generated Graph with Edge Weights")
plt.show()
```
