```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a 2x2 matrix
A = np.array([[3, 1], [1, 3]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
eigenvalues, eigenvectors
```

```python
A @ eigenvectors[:, 0]
```

```python
eigenvalues[0] * eigenvectors[:, 0]
```

```python
# Set up the plot
fig, ax = plt.subplots()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axhline(0, color="grey", lw=0.5)
ax.axvline(0, color="grey", lw=0.5)

# Plot original basis vectors
origin = np.array([[0, 0], [0, 0]])
ax.quiver(
    *origin,
    [1, 0],
    [0, 1],
    color=["blue", "blue"],
    scale=1,
    scale_units="xy",
    angles="xy",
    label="Basis Vectors",
)

# Transform and plot basis vectors under matrix A
transformed_basis = A @ np.array([[1, 0], [0, 1]])
ax.quiver(
    *origin,
    transformed_basis[0],
    transformed_basis[1],
    color=["orange", "orange"],
    scale=1,
    scale_units="xy",
    angles="xy",
    label="Transformed Basis Vectors",
)

# Plot eigenvectors scaled by eigenvalues
for i in range(len(eigenvalues)):
    eig_vec = eigenvectors[:, i] * eigenvalues[i]
    ax.plot(
        [0, eig_vec[0]],
        [0, eig_vec[1]],
        color="red",
        linestyle="--",
        linewidth=2,
        label=f"Eigenvector {i+1}",
    )

# Label and legend
ax.legend(loc="upper left")
ax.set_title("2x2 Matrix Transformation with Eigenvalues and Eigenvectors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid()
plt.show()
```
