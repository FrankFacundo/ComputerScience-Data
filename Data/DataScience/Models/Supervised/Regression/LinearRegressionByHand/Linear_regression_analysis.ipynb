# Linear regression model - Frank Facundo

In first part (Steps 1–8) of this tutorial we consider the following fixed-design one-dimensional $(p = 1)$ linear regression model:

$$
Y_i = \beta _0 + \beta _1 x_i + \epsilon_i,
\qquad \epsilon_i \sim \mathcal{N}(0,\sigma) i.i.d.,
\qquad i=1,...,n
$$

Being a particular but simply interpretable case it facilitates intuitive understanding and enables easy two-dimensional visualization. (Later, in Steps 9–13, we regard a higher-dimensional case.) ML estimators
for $β̂_0$ and $β̂_1$ are:

$$\hat{\beta}_0 = \overline{Y} - \hat{\beta_1}\overline{x}\qquad \text{and} \qquad \hat{\beta_1} = \dfrac{\sum^n_{i=1}(x_i-\bar{x})(Y_i-\overline{Y})}{\sum^n_{i=1}(x_i-\overline{x})^2}$$

with their respective variances being:

$$\mathbb{V}[\hat{\beta_0}] = \hat\sigma^2 \left(\frac{1}{n}+\frac{\overline{x}^2}{\sum^n_{i=1}(x_i-\overline{x})^2}\right) \qquad \text{and} \qquad \mathbb{V}[\hat{\beta_1}] = \frac{1}{n-2}\left( \frac{\hat\sigma^2}{\sum^n_{i=1}(x_i-\overline{x})^2} \right)$$

where $\hat{\sigma}^2$ is the unbiased estimator for $σ^2$ :

$$\hat\sigma^2 = \frac{1}{n-2} \sum^n_{i=1}\left(Y_i - (\hat\beta_0+\hat\beta_1x_i) \right)^2$$

Further, one can show that for any $j = 0, ..., p$ it holds

$$\frac{\hat\beta_j}{\mathbb{V}[\hat\beta_j]} 
\notin 
\left[ 
 -t^{\left(n-(p+1)\right)}_{1-\frac{\alpha}{2}} ; 
  t^{\left(n-(p+1)\right)}_{1-\frac{\alpha}{2}} 
\right],$$

where $t^{\left(n-(p+1)\right)}_{1-\frac{\alpha}{2}}$ is the quantile of the Student-t distribution with $(n-(p+1))$ degrees of freedom at the level $1-\frac{\alpha}{2}$. For a statistical test, the *p*-value is the probability that, under $\mathcal{H}_0$, the test statistic takes
the value at least as extreme as its observed value.

Using the estimated coefficients one obtains the regression line $( \hat\beta_0 + \hat\beta_1 x)$, which is the mean of $Y$ for any $x \in \mathbb{R}$. This is called the point prediction. Usually, one is also interesting in the inferential information, *i.e.* in a confidence interval on $ \hat\beta_0 + \hat\beta_1 x$ and on $Y$ (the last one is often called prediction interval as it reflect the variance of $Y$ as well). One can show that

$$
\frac
{( \hat\beta_0 + \hat\beta_1 x) - ( \beta_0 + \beta_1 x)}
{
	\hat\sigma 
	\sqrt{
		\frac
			{(x-\overline{x})^2}
			{\sum^n_{i=1}(x_i-\overline{x})^2}
		}
}
\sim t(n-2)

\qquad
\text{and}
\qquad

\frac
{Y - ( \hat\beta_0 + \hat\beta_1 x)}
{
	\hat\sigma 
	\sqrt{
		1 + \frac{1}{n} +
		\frac
			{(x-\overline{x})^2}
			{\sum^n_{i=1}(x_i-\overline{x})^2}
		}
}
\sim t(n-2)
$$

This allows for construction of confidence and prediction intervals (at the level $\alpha$):

$$
\text{CI}(x) = 
\hat\beta_0 + 
\hat\beta_1 x \quad \pm 
\quad 
t^{\left(n-2\right)}_{1-\frac{\alpha}{2}}*
\hat\sigma *
\sqrt{
	\frac{1}{n} +
	\frac
		{(x-\overline{x})^2}
		{\sum^n_{i=1}(x_i-\overline{x})^2}
	}
$$

$$
\text{PI}(x) = 
\hat\beta_0 + 
\hat\beta_1 x \quad \pm 
\quad 
t^{\left(n-2\right)}_{1-\frac{\alpha}{2}}*
\hat\sigma *
\sqrt{
	1 + \frac{1}{n} +
	\frac
		{(x-\overline{x})^2}
		{\sum^n_{i=1}(x_i-\overline{x})^2}
	}
$$

The second part (Steps 9–13) of the tutorial deals with the general linear regression model (in the
obvious notation):

$$
Y_i = \beta_0 + \beta_1 x_{x,1} + ... + \beta_p x_{i,p} + \epsilon_i, 
\qquad \epsilon_i \sim \mathcal{N}(0,\sigma) \text{  }i.i.d., 
\qquad i=1,...,n$$

Here, we shortly generalize those of the above mentioned material relevant for performing this tutorial.
Denote the design matrix 
$X = (x_1, ..., x_n)^T$
with 
$x_i = (1, x_{i,1} , ..., x_{i,p})^T$
and let 
$C = \hat\sigma^2 (X^TX)^{-1} $
where

$$\hat\sigma^2 = 
\frac
{1}
{n-(p+1)}
\sum^n_{i=1}
	\left(
		Y_i - 
		(
			\hat\beta_0 + 
			\hat\beta_1 x_{i,1} + 
			... + 
			\hat\beta_p x_{i,p}
		) 
	\right)^2
$$

Then one can show that

$$\mathbb{V}[\hat{\beta_0}] = 
C_{jj}
$$

$$
\text{CI}(x) = 
\hat\beta_0 + 
\hat\beta_1 x_{.,p} + 
... + 
\hat\beta_p x_{.,p}
\quad \pm 
\quad 
t^{\left( n-(p+1) \right)}_{1-\frac{\alpha}{2}}*
\sqrt{
	\hat\sigma^2
	x.^T(X^TX)^{-1}x.
	}
$$

$$
\text{PI}(x) = 
\hat\beta_0 + 
\hat\beta_1 x_{.,p} + 
... + 
\hat\beta_p x_{.,p}
\quad \pm 
\quad 
t^{\left( n-(p+1) \right)}_{1-\frac{\alpha}{2}}*
\sqrt{
	\hat\sigma^2
	(1+x.^T(X^TX)^{-1}x.)
	}
$$

## Exercice 3

For the linear regression of Investment on GNP, estimate the intercept and the slope and their
standard deviations, as well as the determination coefficient for the ordinary least squares. Output
them in a readable form.

We compute a linear regression of Investment on GNP.
In order to do that we estimate the intercept and the slope. Then we compute their standard deviations, as well as the determination coefficient for the ordinary least squares.

```python
# which log? 2? or e? or 10? -> any log is correct because it will work with data after 
# calculate the logarithm operation. In this case it use ln -> np.log

# the regression is based on the results

# X will be thereafter a the log of gnp data
gnp_log = np.log(df['gnp'])
X = gnp_log
# Y will be thereafter a the log of invest data
invest_log = np.log(df['invest'])
Y = invest_log

n = Y.size
X_mean = np.mean(X)
Y_mean = np.mean(Y)
X_mean_center = X - X_mean
Y_mean_center = Y - Y_mean
X_emp_var = np.sum(X_mean_center ** 2)

beta_1 = np.dot(X_mean_center, Y_mean_center) / X_emp_var
beta_0 = Y_mean - beta_1 * X_mean

Y_emp_var = (1 / (n - 2)) * np.sum((Y - (beta_0 + beta_1 * X)) ** 2)

beta_0_sd = np.sqrt(Y_emp_var * ((1 / n) + (X_mean ** 2 / X_emp_var)))

beta_1_sd = np.sqrt(Y_emp_var / X_emp_var)

Y_estimate = beta_0 + beta_1 * X
R2 = np.sum((Y_estimate - Y_mean) ** 2) / np.sum(Y_mean_center ** 2)

print(f"The value of the intercept is {str(beta_0)}")
print(f"The value of the slope is {str(beta_1)}")
print(f"The value of the standard deviation of the intercept is {str(beta_0_sd)}")
print(f"The value of the standard deviation of the slope is {str(beta_1_sd)}")
print(f"The value of the determination coefficient is {str(R2)}")
```

## Exercice 4

Test the significance of the slope using the Student-t test. Report the value of the test statistic and the *p*-value.

It test the significance of the slope using the Student-t test. It compute the value of the test statistic and
the p-value.

Reminder of p-value : 
* If p-value <= significance level (typically 5% or 1%) -> statistically significant.
* If p-value >  significance level (typically 5% or 1%) -> statistically non significant.
* The p-value is the probability of making a mistake by rejecting $H_0$ in other words is the probability of wrongly reject $H_0$
* The p-value is the limit empirical value of $\alpha$

Reminder of other concepts in statistics : 
* By definition, the $1 - \alpha$ level is the probability of rejecting the null hypothesis when the null hypothesis is true. In other words it’s the probability of making a wrong decision.
* $\alpha$ often large for more possible values of the random variable
* If the level $\alpha$ is 1 all values are possible, for example, is $X \sim \mathcal{N}(\mu,\,\sigma^{2})\, X$ can take all real numbers.
* If the level $\alpha$ is 0 no value is possible (impossible to have the value of null hypothesis because the measurement of a singleton is zero).
* It is say that a test is statistically significant if you reject and it is say it is not statistically significant if you do not reject because it can not claim that $H_0$ is correct so it does not conclude anything. 
* p-value depends on the data.

The text above is to be check because $\alpha$ usually is little and the significance level that is $1-\alpha$ is large, in this way all that is said above is correct if we "inverse" 

Let $\widehat {\beta } $ be an estimator of parameter $\beta$ in some statistical model. Then a t-statistic for this parameter is any quantity of the form
$${\displaystyle t_{\widehat {\beta }}={\frac {{\widehat {\beta }}-\beta _{0}}{\operatorname {s.e.} ({\widehat {\beta }})}}}$$

Under the assumption $H_0 = Beta_j = 0$, we have the statistic test $T$ :

```python
T = beta_1 / beta_1_sd
print(f"The statistic test for the slope is {str(T)}")
```

The p-value is the probability of X is greater than T (see TP sheet).

```python
p_value = stats.t.sf(T, n - 2)
print(f"The p-value is {str(p_value)}")
```

As the p-value is almost zero, we can reject the test H0 and accept H1.
So the slope is statistically significant.

## Exercice 9

For the linear regression of Investment on GNP and Interest, compute the associated Gram matrix.
Is it of full rank?

For the linear regression of Investment on GNP and Interest, it computes the associated Gram matrix.

```python
investment = df['invest'].to_numpy()
gnp = df['gnp'].to_numpy()
interest = df['interest'].to_numpy()

x = np.transpose(np.array([np.ones(n), np.log(gnp), interest]))
x_transpose = np.transpose(x)
y = np.log(investment)

Gram = (x_transpose @ x) / n
Gram_inv = np.linalg.inv(Gram)
Gram_eig = np.linalg.eig(Gram_inv)
print(f"The eigenvalues are :{Gram_eig[0]},\nso we can deduce that \"x\" "
      f"is full rank because eigenvalues are different of zero.")
```

For the linear regression of Investment on GNP and Interest, it estimates the three regression co-efficients and their standard deviations, as well as the determination coefficient for the ordinary least squares. Additionally, it reports test significance of each coefficient using the Student-t test. Finally it reports the regression coefficients, corresponding to them p-values, and the determination coefficient in a readable form. Discuss significance of the estimated regression coefficients.

```python
xtx = x_transpose @ x

Theta = np.linalg.inv(xtx) @ (x_transpose @ y)
B0 = Theta[0]
B1 = Theta[1]
B2 = Theta[2]

print(f"\nThe three regression coefficients are {Theta}")

y_emp_var = np.sum((y - (x @ Theta)) ** 2) / (n + 2)
Theta_variance = y_emp_var * np.linalg.inv(xtx)

B0_sd = np.sqrt(Theta_variance[0][0])
B1_sd = np.sqrt(Theta_variance[1][1])
B2_sd = np.sqrt(Theta_variance[2][2])

print(f"\nThe value of the standard deviation of the intercept is {B0_sd}")
print(f"The value of the standard deviation of beta 1 is {B1_sd}")
print(f"The value of the standard deviation of beta 2 is {B2_sd}")

R_2 = np.sum((x @ Theta - np.mean(y)) ** 2) / np.sum((y - np.mean(y)) ** 2)

T1 = B1 / B1_sd
T2 = B2 / B2_sd

print(f'Statistic test for B1 {T1}')
print(f'Statistic test for B2 {T2}')

p_value_B1 = stats.t.sf(T1, n - 3)
p_value_B2 = stats.t.sf(T2, n - 3)
print(f"\nThe p-value for beta 1 is {str(p_value_B1)}")
print(f"The p-value for beta 2 is {str(p_value_B2)}")
```
