```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rates = {
    0: [0, 10777],
    11: [10778, 27478],
    30: [27479, 78570],
    41: [78571, 168994]
}
```

```python
# brut_wages reducted by 20% are taxable_wages
taxable_wages = 33343
```

```python
def compute_taxes(taxable_wages):
    taxe = 0
    for percentage, fraction_wage in rates.items():
        if taxable_wages > fraction_wage[0]:
            upper_limit = min(fraction_wage[1], taxable_wages)
            lower_limit = fraction_wage[0]
            taxe += (percentage/100) * (upper_limit - lower_limit)
    return taxe
```

```python
compute_taxes(taxable_wages)
```

```python
wages = [num for num in range(30000, 100001)]
# wages
```

```python
taxes = [compute_taxes(wage) for wage in wages]
# taxes
```

```python
plt.plot(wages, taxes)
plt.grid()
plt.xlabel('Wages')
plt.ylabel('Taxes')
```

```python
plt.plot(wages, np.array(taxes)/np.array(wages))
plt.grid()
plt.xlabel('Wages')
plt.ylabel('Taxes')
```
