```python
# read_and_plot_gml.py
import networkx as nx
import matplotlib.pyplot as plt

# --- 1. Load the graph ------------------------------------------------------
gml_path = (
    "/home/frank/code/ComputerScience-Data/Scripts/python/crawl/files/web_graph.gml"
)
G = nx.read_gml(gml_path, label="id")  # or label=None if you want 0n labels

print(f"Loaded {G.number_of_nodes()} nodes / {G.number_of_edges()} edges")

# --- 2. Choose a layout -----------------------------------------------------
# A spring layout is often the easiest to start with
pos = nx.spring_layout(G, seed=42)  # deterministic layout

# --- 3. Draw ----------------------------------------------------------------
plt.figure(figsize=(12, 8))
nx.draw_networkx(
    G,
    pos=pos,
    node_size=60,
    font_size=5,
    arrowsize=10,
    with_labels=False,  # turn on if you really want the labels
)
plt.axis("off")
plt.tight_layout()
plt.show()  # or plt.savefig("web_graph_from_gml.png", dpi=300)
```
