```python
def get_classification(min_songs_by_cluster=10):
    songs = list(mm.songs.values())
    song_classification = {}
    nb_clusters = 0
    for genre in genre_possibility_counts:
        if genre[1] >= min_songs_by_cluster:
            nb_clusters += 1
            genre = genre[0]
            for song in list(songs):
                if genre in song.genres:
                    song_classification[song.name] = genre
                    songs.remove(song)
    unclassified = songs
    return song_classification, unclassified, nb_clusters
```

```python
nb_songs = len(list(mm.songs.values()))
nb_songs
```

```python
list_nb_clusters_by_min_song = []
nb_classified_songs = []
result = []
for i in range(55):
    _, unclassified, nb_clusters = get_classification(i)
    result.append((i, nb_songs-len(unclassified), nb_clusters))
```

```python
nb_clusters = [nb_cluster[2] for nb_cluster in result]
nb_classified_songs = [nb_cluster[1] for nb_cluster in result]
```
