```python
import pickle
from pprint import pprint

from my_music import Song
```

```python
# Load the variable from the file
with open("song_list.pkl", "rb") as file:
    loaded_song_list = pickle.load(file)

print(len(loaded_song_list))
# print(loaded_song_list)
```

```python
import matplotlib.pyplot as plt

# Count the number of songs for each top genre
genre_counts = {}
for song in loaded_song_list.values():
    genre = song.top_genre
    if genre in genre_counts:
        genre_counts[genre] += 1
    else:
        genre_counts[genre] = 1

# Sort the genre counts in decreasing order
sorted_genre_counts = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)

# Extract the genres and counts for the bar chart
total_genres = len(sorted_genre_counts)
print(f"Total genres: {total_genres}")
bottom_limit = total_genres - 171
upper_limit = total_genres - 160
genres = [genre for genre, count in sorted_genre_counts[bottom_limit:upper_limit]]  # Limit the x axis with the first 20 elements
counts = [count for genre, count in sorted_genre_counts[bottom_limit:upper_limit]]  # Limit the x axis with the first 20 elements

# Create the bar chart
plt.bar(genres, counts)
plt.xlabel('Top Genre')
plt.ylabel('Number of Songs')
plt.title('Number of Songs per Top Genre')
plt.xticks(rotation=90)
plt.show()
```

```python
def get_songs_by_top_genre(top_genre, search_words=False):
    songs = []
    if search_words:
        for song in loaded_song_list.values():
            if top_genre.lower() in song.top_genre.lower():
                songs.append(song)
    else:
        for song in loaded_song_list.values():
            if top_genre.lower() == song.top_genre.lower():
                songs.append(song)
    return songs
```

```python
def get_songs_by_name(name):
    songs = []
    for song in loaded_song_list.values():
        if name.lower() in song.name.lower():
            songs.append(song)
    return songs
```

```python
def get_songs_by_artist(artist):
    songs = []
    for song in loaded_song_list.values():
        for artist_list in song.artists.keys():
            if artist.lower() in artist_list.lower():
                songs.append(song)
    return songs
```
